```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("groceries.csv")
```

```python
transactions=[]
for i in range(0,9835):
    transactions.append([str(df.iloc[i,j]) for j in range(1,33 ) if str(df.iloc[i,j])!='nan'])
```

```python
transactions[0:3]
```

```python
from apyori import apriori
```

```python
rules=apriori(transactions,min_support=0.003,min_confidence=0.2,min_lift=3,max_length=4)
```

```python
rules=list(rules)
```

```python
rules[0]
```

```python
for i in rules:
    print(i.ordered_statistics[0].items_base)
    print(i.ordered_statistics[0].items_add)
    print("confidence",i.ordered_statistics[0].confidence)
    print("lift",i.ordered_statistics[0].lift)
    print('======')
```

```python
data = pd.DataFrame(columns = ['Items add', 'items base', 'confidence'])
```

```python
 
for i in range(len(rules)):
    ib = rules[i].ordered_statistics[0].items_base
    ia = rules[i].ordered_statistics[0].items_add
    c = rules[i].ordered_statistics[0].confidence
    data.loc[i] = [ia, ib, c]
```
